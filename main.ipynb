## Ethnicity  
*  Ethnicity refers to a group of people who share a common culture. It is sometimes associated with their national origin.
*  For example, someone may be Asian, but ethnically Vietnamese.    
## Radiation_Exposure
*  it's a common treatment option for many types of cancer. 
## Family_history
* A family history of cancer is a record of cancer cases in close relatives.
## Iodine_Deficiency	
*  Iodine deficiency is a condition that occurs when your body doesn't have enough iodine
## Obesity 
* Obesity is a chronic disease that occurs when there is too much body fat.
## TSH_Level 
* A higher TSH level can indicate a higher risk of thyroid malignancy.
## T3_level 
* A T3 test measures the level of triiodothyronine in the blood
## T4_level 
* In most cases, T4 levels (thyroxine) in a person with thyroid cancer will be normal

```python
##Featuer selection Function 
def FeatuerSelection(numer_of_rows):
    sfc = SequentialFeatureSelector(dt ,  n_features_to_select=numer_of_rows , direction= 'forward' , scoring='accuracy' )
    sfc.fit(X_train , y_train)

    print("FeatuerNames -->" , sfc.get_feature_names_out())

    df_featuers = df[['Age', 'Gender', 'Family_History', 'Radiation_Exposure',
       'Iodine_Deficiency', 'Smoking', 'Obesity', 'Diabetes', 'T4_Level']]
    
    X_train_fe , X_test_fe , y_train_fe , y_test_fe = train_test_split(df_featuers , df['Thyroid_Cancer_Risk'] , test_size=0.3 , random_state=42)

    dt.fit(X_train_fe , y_train_fe)
    y_pred_fe = dt.predict(X_test_fe)
    print(accuracy_score(y_test_fe , y_pred_fe))
    return 0
    
```

## n_estimators: the number of decision trees in the forest.

```python
##traing ans testing accuracy gragphh
from sklearn.ensemble import RandomForestClassifier
import plotly.express as ex
max_depths  = np.arange(1,11)

train_accuracies = []
test_accuracies = []

for i in max_depths:
    # print(f"n_estimators --> {i}")
    # print(f"max_depth ==> {i}")
    rc = RandomForestClassifier(n_estimators= i , criterion='entropy' , max_depth = i , min_samples_split=2)
    rc.fit(X_traind_df , y_train)

    # y_pred_fn = dt.predict(X_test_encoded)
    # train_accuracies.append(accuracy_score(y_test , y_pred_fn))
    
    # test_accuracies.append(accuracy_score(y_train , dt.predict(X_train_encoded))) 
    train_accuracies.append(rc.score(X_traind_df, y_train))
    test_accuracies.append(rc.score(X_test_encoded, y_test))

print("train_accuracies -- >" , train_accuracies)
print("test_accuracies -->" , test_accuracies)

plt.plot(max_depths , train_accuracies , label = "Train-Accuracy" , marker = "*")
plt.plot(max_depths , test_accuracies  , label = "Test-Accuracy")
```
